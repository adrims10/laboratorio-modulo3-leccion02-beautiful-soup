```python
def obtener_datos_vazquez(total_paginas=200):
    secciones_totales = []
    nombres_totales = []
    cat_totales = []
    imagenes_totales = []

    for page in range(1, total_paginas + 1):
        url_vazquez = f'https://atrezzovazquez.es/shop.php?search_type=-1&search_terms=&limit=48&page={page}'
        res_vazquez = requests.get(url_vazquez)
        
        if res_vazquez.status_code != 200:
            print(f"Error al acceder a la página {page}: {res_vazquez.status_code}")
            continue
        
        sopa_vazquez = BeautifulSoup(res_vazquez.content, 'html.parser')  
        
     
        seccion = sopa_vazquez.findAll('a', {'class': "tag"})
        secciones = [sec.getText() for sec in seccion]
        secciones_totales.extend(secciones)

        
        nombres = sopa_vazquez.findAll('a', {'class': "title"})
        nombres_totales.extend([nom.getText() for nom in nombres])

        
        cat = sopa_vazquez.findAll('div', {'class': "cat-sec-box"})
        cat_totales.extend([cate.getText() for cate in cat])

       
        lista_imagen = sopa_vazquez.findAll("div", {"class": "product-image"})
        for imagen_div in lista_imagen:
            img_tag = imagen_div.find("img")  # Buscar la etiqueta <img> dentro del div
            if img_tag:
                imagen_src = img_tag.get('src')
                # Asegúrate de construir la URL correctamente
                imagenes_totales.append(f"https://atrezzovazquez.es/{imagen_src}")

    
    max_len = max(len(secciones_totales), len(nombres_totales), len(cat_totales), len(imagenes_totales))

    secciones_totales += [None] * (max_len - len(secciones_totales))
    nombres_totales += [None] * (max_len - len(nombres_totales))
    cat_totales += [None] * (max_len - len(cat_totales))
    imagenes_totales += [None] * (max_len - len(imagenes_totales))

    
    datos = {
        'Sección': secciones_totales,
        'Nombre': nombres_totales,
        'Categoría': cat_totales,
        'Imagen': imagenes_totales
    }

    df = pd.DataFrame(datos)
    return df
```

```python
df_vazquez = obtener_datos_vazquez()
df_vazquez
```
